## Section 1. Download tracks for all 5 playlists

### Two functions will be used to acquire the necessary information for all 5 playlists:
  1. **download_tracks**(playlist, user_id = USER_ID)
      <br> ex: `drumcode_tracks = download_tracks(drumcode)`
      - Get track_ids for a playlist, fit info into dataframe, drop tracks without mp3 file
      - Get audio features using track_ids, create a second dataframe, drop unnecessary columns
      - Merge both dataframes into one
        
  2. **save_and_download_mp3**(playlist_df, playlist_name)
      <br> ex: `save_and_download_mp3(drumcode_tracks, drumcode_name)`
      - Save dataframe in the "datasets" folder
      - Download mp3 files to the downloads/playlist folder (create empty playlist folders ahead of time)

```python
def download_tracks(playlist, user_id = USER_ID):
    track_id = []
    track_name = []
    mp3 = []
    track_length = []
    track_pop = []
    album_date = []
    artist_id = []
    artist_name = []
    None_counter = 0 
    
#DOWNLOAD TRACKS, append track info to empty lists   
    for j in range(0, 3000, 100):
        results = sp.user_playlist_tracks(USER_ID, playlist, offset=j)
        for i, t in enumerate(results['items']): 
            track_id.append(t['track']['id'])
            track_name.append(t['track']['name'])
            mp3.append(t['track']['preview_url'])
            if t['track']['preview_url'] == None:
                None_counter += 1          #don't want to keep these tracks without an mp3 link 
            track_length.append(t['track']['duration_ms'])
            track_pop.append(t['track']['popularity'])
            album_date.append(t['track']['album']['release_date'])
            for a, b in enumerate(t['track']['artists']):
                if a == 0:      #only prints the FIRST artist  (a=0(1 artist);  a=1(2 artists), etc)
                    artist_id.append(b['id'])
                    artist_name.append(b['name'])
    
#put all the new lists into a dataframe
    df_tracks = pd.DataFrame({'track_id':track_id,"track_name": track_name, "mp3": mp3, "track_length": track_length, 
                             "track_pop": track_pop, "album_date": album_date, "artist_id": artist_id, 
                              "artist_name": artist_name})

    print('Total number of tracks in playlist:', len(df_tracks))
    print('Number of tracks without mp3 link:',None_counter)
    print('Number of usable tracks:', int(len(df_tracks))-None_counter)

# drop any columns that have mp3 = None, and reset index
    df_tracks = df_tracks.dropna(subset = ["mp3"])
    df_tracks.reset_index(drop=True, inplace=True)
    print('Shape of ', playlist, '=', df_tracks.shape)
    print("------------------------------------------")
#     return df_tracks
#--------------------------------------------------------------------------------
# GET AUDIO FEATURES
    rows = []          #empty list to append each track features to 
    batchsize = 100    #max number of track ids we're allowed to submit per query
    None_counter = 0   #count if there are any songs without any audio features

    for i in range(0,len(df_tracks['track_id']), batchsize):    #step = batchsize
        batch = df_tracks['track_id'][i:i+batchsize]
        feature_results = sp.audio_features(batch)      #query the audio features endpoint
        for i, t in enumerate(feature_results):
            if t == None:
                None_counter += 1         
            else:
                rows.append(t)

    print('Number of tracks where no audio features were available:',None_counter)
    print('Number of usable tracks:', len(rows))

# create a dataframe of these audio features
    df_audio_features = pd.DataFrame.from_dict(rows, orient='columns')
    # Drop unnecessary columns
    columns_to_drop = ['analysis_url','track_href','type','uri']
    df_audio_features.drop(columns_to_drop, axis=1,inplace=True)
    # rename the "id" column so it matches the other dataframe
    df_audio_features.rename(columns={'id': 'track_id'}, inplace=True)
    print("Audio_features shape =", df_audio_features.shape)

# merge both dataframes
    df = df_tracks.join(df_audio_features.set_index("track_id"), on = ('track_id'))
    print("Shape of the merged dataset:", df.shape)
    return df
```

```python
def save_and_download_mp3(playlist_df, playlist_name):
    playlist_df.to_csv("./datasets/"+playlist_name+".csv") #save df in datasets folder
    print("CSV saved")
    print("mp3 download started...")
    count = 0
    for index, row in playlist_df.iterrows():  #download mp3 files to local drive
        count += 1
        url = row['mp3']
        track_name = row['track_id']
        urllib.request.urlretrieve(url, './downloads/'+playlist_name+'/'+track_name+'.mp3')
    print(count, "mp3 tracks downloaded")
    return
```

## Section 2. Convert mp3 files to spectrograms

```python
song_folder = ('./downloads/')


idx_to_genre = []
genre_to_idx = {}
song_specs = []
genres = []

for genre in os.listdir(song_folder):
    if genre != ".DS_Store":
        genre_to_idx[genre] = len(genre_to_idx)
        idx_to_genre.append(genre)
        genre_folder = os.path.join(song_folder, genre) 
        for song in os.listdir(genre_folder):
            if song.endswith('.mp3'):
                mp3_file  = os.path.join(genre_folder, song)

                signal, sr = librosa.load(mp3_file)
                S = librosa.feature.melspectrogram(signal, sr=sr, n_mels=128).T[:1280,]
#                 log_S = librosa.power_to_db(S, ref=np.max)
                log_S = librosa.power_to_db(S)
                if S.shape[0] == 1280:  #some songs were found to be less than 30 sec clip
                    song_specs.append(log_S)
                    genres.append(genre_to_idx[genre])
                
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel') 
plt.colorbar()
```

## Section 3. 1D Neural Network

```python
import keras
from keras import regularizers
from keras.models import Sequential, Model
from keras.layers import Input, LSTM, Dense, Dropout, Flatten, Activation
from keras.layers.merge import concatenate
from keras.layers.convolutional import Conv1D, Conv2D
from keras.layers.pooling import MaxPooling1D, GlobalMaxPooling1D, GlobalAveragePooling1D, MaxPooling2D, GlobalMaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
from sklearn.model_selection import train_test_split

from timeit import default_timer as timer
```

```python
#one hot encodes my categories
genres_one_hot = keras.utils.to_categorical(genres_file, num_classes=len(genre_to_idx)) 
print(genres_one_hot.shape)
print(genres_one_hot)
```

```python
# STEP 2: train-test-split
X_train, X_test, y_train, y_test = train_test_split(
#                                                     np.array(song_specs_file),
                                                    song_specs_file,
#                                                     np.array(genres_one_hot),
                                                    genres_one_hot,
                                                    test_size=0.1,
                                                    random_state = 42,
                                                    stratify=genres
                                                   )

print(X_train.shape)
print(y_train.shape)
```

```python
y_train
```

```python
def split_10(x, y):
    s = x.shape
    s = (s[0] * 10, s[1] // 10, s[2])
    return x.reshape(s), np.repeat(y, 10, axis=0)

X_test, y_test = split_10(X_test, y_test)
X_train, y_train = split_10(X_train, y_train)

print(X_train.shape)
print(y_train.shape)
```

```python
model = Sequential()
model.add(Conv1D(filters = 64, kernel_size = 4, 
                 activation = 'relu', 
                 input_shape = (128, 128), 
                 kernel_regularizer=regularizers.l2(0.001)))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size = 2, strides = 2))
# model.add(Dropout(0.3))

model.add(Conv1D(128, kernel_size = 4, activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size = 2, strides = 2))
# model.add(Dropout(0.5))

model.add(Conv1D(256, kernel_size = 4, activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size = 2, strides = 2))

model.add(GlobalMaxPooling1D())
model.add(Dropout(0.3))

model.add(Dense(256, activation = 'relu'))
model.add(Dropout(0.3))

model.add(Dense(5, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', 
              metrics = ['accuracy'])
              
model.summary()
```

```python
start = timer()

results = model.fit(X_train, y_train, epochs = 20, validation_data=(X_test, y_test), 
                   batch_size = 5000)

end = timer()
print((end - start)/60)
```

```python
plt.figure(figsize = (10, 3))
plt.subplot(121)
plt.plot(results.history['loss'], label = 'Train Loss')
plt.plot(results.history['val_loss'], label = 'Test Loss')
plt.subplot(122)
plt.plot(results.history['acc'], label = 'Train Accuracy')
plt.plot(results.history['val_acc'], label = 'Test Accuracy')
plt.legend()

print("Train accuracy =", round(results.history['acc'][0], 2), round(results.history['acc'][10], 2), 
      round(results.history['acc'][-1], 2))
print("Test accuracy  =", round(results.history['val_acc'][0], 2), round(results.history['val_acc'][10], 2),
      round(results.history['val_acc'][-1], 2))

print("Train loss =", round(results.history['loss'][0], 2), round(results.history['loss'][10], 2), 
      round(results.history['loss'][-1], 2))
print("Test loss =", round(results.history['val_loss'][0], 2), round(results.history['val_loss'][10], 2), 
      round(results.history['val_loss'][-1], 2))
```

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
y_test = np.argmax(y_test, axis=1)
print(confusion_matrix(y_test, y_pred))
```
